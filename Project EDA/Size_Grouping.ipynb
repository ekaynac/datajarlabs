```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_cleaned_thresholded.csv")
df
```

```python
print("Max data:{}".format(max(df.groupby(by="location")["new_cases"].count())))
print("Min data:{}".format(min(df.groupby(by="location")["new_cases"].count())))
```

```python
df.groupby(by="location")["new_cases"].count().sort_values()
plt.figure(figsize=(15,15))
plt.scatter(df.groupby(by="location")["new_cases"].count().sort_values(),df.groupby(by="location")["new_cases"].count().sort_values().index)
plt.axvline(55,color="red")
plt.axvline(80,color="red")
plt.show()
```
