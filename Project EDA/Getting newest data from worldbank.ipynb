```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hns = pd.read_csv("Health_Nutrition_and_Statistics.csv")
wdi = pd.read_csv("world_development_indicators.csv")
```

```python
hns.head()
```

```python
for cname in hns:
    if cname == "Year":
        continue
    elif cname == "Country Name":
        continue
    elif cname == "Country Code":
        continue
    elif cname == "Time Code":
        continue
    else:
        hns[cname] = pd.to_numeric(hns[cname], errors='coerce')
```

```python
for cname in wdi:
    if cname == "Year":
        continue
    elif cname == "Country Name":
        continue
    elif cname == "Country Code":
        continue
    elif cname == "Time Code":
        continue
    else:
        wdi[cname] = pd.to_numeric(wdi[cname], errors='coerce')
```

```python
wdi.head()
```

```python
wdi.info()
```

```python
i=0
enough_info_wdi = list()
not_enough_info_wdi = list()
threshold_wdi = 150

for p in wdi:
    i=0
    for j in range(len(wdi.groupby(by="Country Name").count())):
        if p == "Country Name":
            continue
        
        elif wdi.groupby(by="Country Name").any()[p][j] == True:
            i+=1
    if i < threshold_wdi:
        not_enough_info_wdi.append(p)
    else:
        enough_info_wdi.append(p)
```

```python
print("How many features in World Development Indicators not have enough data: {} ".format(len(not_enough_info_wdi)))
```

```python
print("How many features in World Development Indicators have enough data: {} ".format(len(enough_info_wdi)))
```

```python
i=0
enough_info_hns = list()
not_enough_info_hns = list()
threshold_hns = 150

for p in hns:
    i=0
    for j in range(len(hns.groupby(by="Country Name").count())):
        if p == "Country Name":
            continue
        
        elif hns.groupby(by="Country Name").any()[p][j] == True:
            i+=1
    if i < threshold_hns:
        not_enough_info_hns.append(p)
    else:
        enough_info_hns.append(p)
```

```python
print("How many features in Health Nutrition and Population not have enough data: {} ".format(len(not_enough_info_hns)))
```

```python
print("How many features in Health Nutrition and Population have enough data: {} ".format(len(enough_info_hns)))
```

```python


for name in not_enough_info_wdi:
    if name == "Country Name":
        pass
    else:
        del wdi[name] 
```

```python
for name in not_enough_info_hns:
    if name == "Country Name":
        pass
    elif name == "Year Code":
        pass   
    else:
        del hns[name] 
```

```python
samefeatures = list()
for wdix in wdi:
    for hnsx in hns:
        if wdix == hnsx:
            samefeatures.append(wdix)
            
samefeatures
```

```python
del samefeatures[1]
del samefeatures[0]

samefeatures
```

```python
for wdiname in wdi:
    for same in samefeatures:
        if wdiname==same:
            del wdi[wdiname]
```
