```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("full_data.csv")
df
```

```python
df_nnzero = df[df.total_cases != 0]
df_nnzero
```

```python
df_nnzero.groupby(by="location").count().mean()["date"]

```

Ask for how many country we have data of more than 30 days?

```python
x = 0
y = 0
threshold = 14
for i in range(len(df_nnzero.groupby(by="location"))):
    if (df_nnzero.groupby(by="location").count() > threshold)["date"][i] == True:
        x += 1
print(" {} country have more than {} day data".format(x,threshold))


for i in range(len(df_nnzero.groupby(by="location"))):
    if (df_nnzero.groupby(by="location").count() > threshold)["date"][i] == False:
        y += 1
print(" {} country have less than {} day data".format(y,threshold))
```

In our model we need to have at least one mounth data for a country to add it to our dataset. Threshold is 
adjustable so we can change it whenever we want.

```python
(df_nnzero.groupby(by="location").count() > threshold)
```

```python
df_nnzero.reset_index(inplace= True)
```

```python
th_country_list = list()
for i in range(len(df_nnzero.groupby(by="location"))):
    if ((df_nnzero.groupby(by="location").count()).date[i] > threshold) == True:
        th_country_list.append((df_nnzero.groupby(by="location").count()).index[i])
        
len(th_country_list)
```

```python
df_thresh = pd.DataFrame()
p=0
for i in range(len(df_nnzero["date"])):
    for country in th_country_list:
        if df_nnzero["location"][i]==country:
            df_thresh= df_thresh.append(df_nnzero[df_nnzero.index == i])

df_thresh.reset_index(inplace = True)

df_thresh
```

Now clean unnecessary columns and location == "world"

```python
del df_thresh["level_0"]
del df_thresh["index"]
del df_thresh["date"]
del df_thresh["population_year"]

df_thresh
```

```python
df_world = df_thresh[df_thresh["location"] =="World"]
df_world.reset_index(inplace=True) 
del df_world["index"]
df_world
```

```python
df_thresh = df_thresh[df_thresh["location"] != "World"]
```
