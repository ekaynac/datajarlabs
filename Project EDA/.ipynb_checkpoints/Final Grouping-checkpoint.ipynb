We need to determine what partion of our data we will be working on. Will we keep all data with total_cases == 1 for several first days, or will we start from threshold of total_cases say 100.
Since our data for any country should have the same lenght for every country if we threshold our data we will have to cut some countries data which Im not very fond of. Let us see.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datacheck15-4-20.csv")
df
```

```python
thresholded =list()
plt.figure(figsize = (15,10))
for threshold in range(50,201,10): 
    df_thresholded = df[df["total_cases"]>threshold]
    thresholded.append(len(df_thresholded.groupby(by="Country Name").count()[df_thresholded.groupby(by="Country Name").count()["continent"]>14]))

plt.scatter(range(50,201,10),thresholded)
plt.axvline(80,color="red")
```

There is a "knee" on the threshold==80. I want to choose 80 as my threshold.

```python
x = len(df_thresholded.groupby(by="Country Name").count()[df_thresholded.groupby(by="Country Name").count()["continent"]>14])
print("We have {} country that have more than 14 days data that at least have 80 total cases.".format(x))
```

It okay for me to work with 53 country rather than 155 :(

```python
df_th = df[df["total_cases"] >= 80]
df_th.reset_index(inplace=True,drop=True)
df_th
```

```python
print("Max data:{}".format(max(df_th.groupby(by="Country Name")["total_cases"].count())))
print("Min data:{}".format(min(df_th.groupby(by="Country Name")["total_cases"].count())))
```

```python
country_cut = df_th.groupby(by="Country Name").count()
country_cut = country_cut[country_cut["total_cases"]>=14]
allowed_countries=country_cut.index
allowed_countries
```

```python
df_th["flag"] = False
for index in range(len(df_th)):
    for countryallowed in allowed_countries:
        if df_th.loc[index ,"Country Name"] == countryallowed:
            df_th.loc[index,"flag"] = True
```

```python
df_th = df_th[df_th["flag"]]
del df_th["Time"]
del df_th["flag"]
del df_th["2 weeks"]
del df_th["4 weeks"]
del df_th["+8 weeks"]
del df_th["Unnamed: 0"]
del df_th["Unnamed: 0.1"]
del df_th["Unnamed: 0.1.1"]
df_th.reset_index(inplace=True,drop=True)
df_th
```

```python
plt.figure(figsize=(15,30))
plt.scatter(df_th.groupby(by="Country Name").count()["total_cases"],df_th.groupby(by="Country Name").count().index)
plt.axvline(14,color="red")
plt.axvline(21,color="purple")
plt.axvline(28,color="blue")
plt.axvline(35,color="green")
plt.show()
```

```python
datacountpercountry = df_th.groupby(by="Country Name").count()["total_cases"] 

datacountpercountry
```

```python
countries2week = datacountpercountry[datacountpercountry>=14].index
countries2week
```

```python
countries3week = datacountpercountry[datacountpercountry>=21].index
countries3week
```

```python
countries4week = datacountpercountry[datacountpercountry>=28].index
countries4week
```

```python
countries5week = datacountpercountry[datacountpercountry>=35].index
countries5week
```

```python
countries6week = datacountpercountry[datacountpercountry>=42].index
countries6week
```

```python
df_th["2week"]= False
df_th["3week"]= False
df_th["4week"]= False
df_th["5week"]= False
df_th["6week"]= False
```

```python
df_th
```

```python
for country in countries2week:
    i=0
    for index in range(len(df_th)):
        if df_th.loc[index,"Country Name"] == country:
            if i <14:
                df_th.loc[index,"2week"] = True
                i = i+1

for country in countries3week:
    i=0
    for index in range(len(df_th)):
        if df_th.loc[index,"Country Name"] == country:
            if i <21:
                df_th.loc[index,"3week"] = True
                i = i+1

for country in countries4week:
    i=0
    for index in range(len(df_th)):
        if df_th.loc[index,"Country Name"] == country:
            if i <28:
                df_th.loc[index,"4week"] = True
                i = i+1

for country in countries5week:
    i=0
    for index in range(len(df_th)):
        if df_th.loc[index,"Country Name"] == country:
            if i <35:
                df_th.loc[index,"5week"] = True
                i = i+1
            
for country in countries6week:
    i=0
    for index in range(len(df_th)):
        if df_th.loc[index,"Country Name"] == country:
            if i <42:
                df_th.loc[index,"6week"] = True
                i = i+1
            
```
