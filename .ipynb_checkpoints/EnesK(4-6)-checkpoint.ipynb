```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("StudentsPerformance.csv")
df
```

```python
plt.figure(figsize=(15,10))
df["all_score_mean"]= (df["math score"] + df["reading score"] + df["writing score"]) /3
plt.subplot(1,3,1)
plt.scatter(df["race/ethnicity"],df["all_score_mean"])
plt.title("ethnicity/score")
plt.subplot(1,3,2)
plt.scatter(df["parental level of education"],df["all_score_mean"])
plt.title("parental degree/score")
plt.subplot(1,3,3)
plt.scatter(df["gender"],df["all_score_mean"])
plt.title("gender/score")
plt.show()
plt.figure(figsize=(18,15))
plt.subplot(2,3,1)
plt.scatter(df["lunch"],df["all_score_mean"])
plt.title("lunch/score")
plt.subplot(2,3,2)
plt.scatter(df["test preparation course"],df["all_score_mean"])
plt.title("prep/score")
plt.show()
```

Students who eat standard lunch tend to be more successful than students who eat free/reduced lunch.

It is obvious that exam performance affected by ethnicity, parental degree and gender.

Students who took preparation school tend to be more successful than students didn't take preparation school.

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(df["math score"],df["reading score"])
plt.subplot(1,2,2)
plt.scatter(df["math score"],df["writing score"])
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(df["reading score"],df["writing score"])
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr)
plt.title("Korelasyon")
plt.show()
```

Writing and reading scores' correlation is the biggest in between the three exam scores.

But the biggest correlation is between all_score_mean and reading score.
