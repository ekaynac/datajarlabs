```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weatherHistory.csv")
df
```

```python
X = df[["Humidity","Wind Speed (km/h)","Wind Bearing (degrees)","Pressure (millibars)"]]
Y = df["Temperature (C)"]
```

```python
lrm = linear_model.LinearRegression()

lrm.fit(X,Y)
```

```python
plt.figure(figsize=(15,10))
plt.plot(range(df.shape[0]),np.sort(Y),color="blue")
plt.plot(range(df.shape[0]),np.sort(lrm.predict(X)),color="red")
plt.show()
```

```python
X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

Check for Gauss Markov assumptions:

1) Mean error is zero:

```python
X = df[["Humidity","Wind Speed (km/h)","Wind Bearing (degrees)","Pressure (millibars)"]]
prediction = lrm.predict(X)
error = Y - prediction

print("Mean Error : {:.15f}".format(np.mean(error)))
```

2) Homoscedasticity

```python
plt.figure(figsize=(15,10), dpi=90)
plt.scatter(prediction, error)
plt.xlabel('Prediction')
plt.ylabel('Error')
plt.axhline(y=0,color="red")
plt.title('Prediction x Error')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(error)
plt.show()
```

```python
from statsmodels.tsa.stattools import acf

acf_data = acf(error,fft=False)

plt.figure(figsize=(15,10))
plt.plot(acf_data[1:])
plt.show()
```

```python
house_features = pd.read_csv("processed_houseprice_features.csv")

del house_features["Unnamed: 0"]

house_features
```

```python
house_prices = pd.read_csv("houseprices_for_processed.csv")

house_prices
```

```python
lrm = linear_model.LinearRegression()

lrm.fit(house_features,house_prices)
```

Gauss Markov Assumptions:

```python
prediction = lrm.predict(house_features)
error = house_prices - prediction

print("Mean Error: ",np.mean(error))
```

```python
plt.figure(figsize=(15,10), dpi=90)
plt.axhline(y=0,color="red")
plt.scatter(prediction, error)
plt.xlabel('Prediction')
plt.ylabel('Error')
plt.title('Error x Prediction')
plt.show()
```

Heteroscesdastisity problem

```python
from statsmodels.tsa.stattools import acf

acf_data = acf(error)

plt.figure(figsize=(15,10))
plt.plot(acf_data[1:])
plt.show()
```
